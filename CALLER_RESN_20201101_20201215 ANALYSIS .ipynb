**convertion of the column 'ID_DATE' from int to DateTime objects.**

```python
df
```

```python
df['ID_DATE'] = df['ID_DATE'].astype(str)
df['ID_DATE'] = pd.to_datetime(df['ID_DATE'], format='%Y%m%d', errors='coerce')
```

```python
df
```

```python
df['Month'] = df['ID_DATE'].apply(lambda time: time.month)
```

```python
df['Day_of_Week'] = df['ID_DATE'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day_of_Week'] = df['Day_of_Week'].map(dmap)
```

```python
dmap = {12:'December',11:'November'}
```

```python
df['Month'] = df['Month'].map(dmap)
```
